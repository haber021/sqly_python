```python
import MySQLdb as db
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set_style(style="darkgrid")# "whitegrid")
%matplotlib inline 
sql_conn = db.connect('localhost','root','','poducts') 
```

```python
query ='SELECT * FROM product;'
```

```python
df = pd.read_sql(query, sql_conn)

df.head(10)
```

```python
df_data = (df[df['Continent'] == 'Asia'])
df_data
```

```python
df_new = df_data[(df_data["Continent"]=="Asia") & (df_data["Region"]=="Southeast Asia")]
df_new
```

```python
df_new = df_data[(df_data["Continent"]=="Asia") & (df_data["Name"]=="Philippines")]
df_new
```

```python
df = df_new.nlargest(5,['SurfaceArea'],'first')
df
```

```python
df_life = df_new.nsmallest(5,['Population'],'first')
df_life
```

```python
per_sur_nonsur = (df_life["Region"].value_counts()/df_life.shape[0]*100).round(2)
per_sur_nonsur
```

```python
per_sur_nonsur = (df_life["Name"].value_counts()/df_life.shape[0]*100).round(2)
per_sur_nonsur
```
